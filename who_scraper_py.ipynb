```python
overview = {
    "Indicator": [
        "Population (2023)",
        "Current health expenditure (% of GDP, 2021)",
        "WHO region",
        "World Bank income level"
    ],
    "Value": [
        786385,
        3.85,
        "South East Asia",
        "Lower-middle income"
    ]
}

df_overview = pd.DataFrame(overview)
df_overview
```

```python
population_data = {
    "Year": [2023, 2050],
    "Population": [786385, 882340]  # +12% projection
}

df_population = pd.DataFrame(population_data)
df_population
```

```python
plt.figure(figsize=(6,4))
plt.plot(df_population["Year"], df_population["Population"], marker="o")
plt.title("Population Trend (Bhutan)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.show()
```

```python
life_expectancy = {
    "Year": [2000, 2021],
    "Bhutan (years)": [66, 74.9],
    "SEA Region (years)": [64.1, 68.4]
}

df_life = pd.DataFrame(life_expectancy)
df_life
```

```python
plt.figure(figsize=(7,4))
plt.plot(df_life["Year"], df_life["Bhutan (years)"], marker="o", label="Bhutan")
plt.plot(df_life["Year"], df_life["SEA Region (years)"], marker="o", label="South-East Asia")
plt.title("Life Expectancy Comparison")
plt.xlabel("Year")
plt.ylabel("Life Expectancy (years)")
plt.legend()
plt.grid(True)
plt.show()
```
